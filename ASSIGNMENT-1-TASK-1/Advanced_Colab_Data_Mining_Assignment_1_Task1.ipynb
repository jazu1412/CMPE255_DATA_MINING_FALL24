```python




def folder_list(folder_id):
    from googleapiclient.discovery import build


    service = build('drive', 'v3')


    query = "'{}' in parents".format(folder_id)

    res = service.files().list(q=query, fields="files(id, name)").execute()


    return [{'id': f['id'], 'name': f['name']} for f in res.get('files', [])]

```

## LOOPED AND FILES ARE DOWNLOADED

```python
def folder_download(folder_id):
    from googleapiclient.discovery import build
    import subprocess

    # Build the Drive API client
    service = build('drive', 'v3')

    # Create the query to find files in the specific folder
    query = "'{}' in parents".format(folder_id)
    res = service.files().list(q=query, fields="files(id)").execute()

    # Iterate over the files and download each one
    for file in res.get('files', []):
        file_id = file['id']
        # Use subprocess to call gdown with the correct file ID
        subprocess.run(['gdown', '--id', file_id, '-q'])

```

#DRAWING WITH SVG

```python
import re
from IPython.display import HTML
from IPython.core.magic import register_cell_magic
```

```python
def var_sub(m):
  v = globals()[m[1]]
  if type(v) != str:
    v = repr(v)
  return v
```

```python
import re
from IPython.core.magic import register_cell_magic
from IPython.display import display, HTML

@register_cell_magic
def javascript(line="", cell=""):

    def var_sub(match):
        var_name = match.group(1)
        return globals().get(var_name, f"{{{{ {var_name} }}}}")

    cell = re.sub(r'\{\{(\w+)\}\}', var_sub, cell)

    m = re.match(r'\w+', line)
    if m:
        globals()[m[0]] = cell
    else:
        display(HTML("""
<div id="drawing"></div>
<script src="https://cdnjs.cloudflare.com/ajax/libs/svg.js/2.6.5/svg.js"></script>
<script>
var draw = SVG('drawing').size('100%', 100)
var y_cursor_pos = 0
function print(x){
  draw.text(x.toString()).move(400, y_cursor_pos)
  y_cursor_pos += 20
}

text = s => draw.text(s).font('anchor', 'middle')
line = (x1, y1, x2, y2) => draw.line(x1, y1, x2, y2).stroke({ width: 1 })
max = Math.max


function drawCircle(x, y, radius) {
    draw.circle(radius * 2).center(x, y).stroke({ width: 1 }).fill('none')
}

""" + cell + """
</script>
"""))
```

## Function to display JSON response

```python
from requests import models, Session, Request
from urllib.parse import urljoin
from IPython.core.magic import register_cell_magic
import re


def render(r):
    text = r.text
    if text[0] in "[{":
        return """
        <script src="https://rawgit.com/caldwell/renderjson/master/renderjson.js"></script>
        <script>
        renderjson.set_show_to_level(1)
        document.body.appendChild(renderjson(%s))
        new ResizeObserver(google.colab.output.resizeIframeToContent).observe(document.body)
        </script>
        """ % text
    else:
        return "<pre>%s</pre>" % text

models.Response._repr_html_ = render
```

##REGISTER ELASTIC SEARCH CELL

```python

# Define the %%es magic function
@register_cell_magic
def es(line=None, cell=""):
    cell = re.sub(r'(?m)^\s*#.*\n?', '', cell)
    line1 = (cell + '\n').find('\n')
    method, path = cell[:line1].split(None, 1)
    body = cell[line1:].strip()
    args = {}
    if body:
        args['data'] = (body + '\n').encode()
        args['headers'] = {'Content-Type': 'application/json'}

    rsp = Session().send(
        Request(method, urljoin('http://localhost:9200', path), **args).prepare())
    return rsp
```

### SQUARE OF A NUMBER -  JS CALLS PYTHON

```python
from IPython.display import JSON
import IPython.display as idisplay
from google.colab import output

def square_number(x):
    result = x ** 2
    return JSON({'result': result})

output.register_callback('notebook.square_number', square_number)

js_code = ''' (async function() {
  const number = prompt("Enter a number to square:");
  if (number !== null) {
    const num = parseFloat(number);
    if (!isNaN(num)) {
      const result = await google.colab.kernel.invokeFunction(
        'notebook.square_number',
        [num],
        {});

      const squaredResult = result.data['application/json'].result;

      const outputText = `The square of ${num} is ${squaredResult}`;
      document.querySelector("#output-area").appendChild(document.createTextNode(outputText));
    } else {
      alert("Please enter a valid number.");
    }
  }
})();
'''

idisplay.display(idisplay.Javascript(js_code))
```

TAKE A SNAP

```python
from IPython.display import HTML, display
from google.colab.output import eval_js
from base64 import b64decode

# Define the HTML and JavaScript for capturing the photo
PHOTO_CAPTURE_HTML = """
<video autoplay width=640 height=480></video>
<button id="capture">Capture Photo</button>
<canvas id="canvas" style="display: none;"></canvas>
<script>
var video = document.querySelector('video');
var canvas = document.getElementById('canvas');
var context = canvas.getContext('2d');
var captureButton = document.getElementById('capture');

// Start the video stream
navigator.mediaDevices.getUserMedia({ video: true })
  .then(stream => video.srcObject = stream);

// Set up capture button
captureButton.onclick = () => {
  // Set canvas dimensions to match video
  canvas.width = video.videoWidth;
  canvas.height = video.videoHeight;

  // Draw video frame to canvas
  context.drawImage(video, 0, 0);

  // Stop the video stream
  video.srcObject.getVideoTracks()[0].stop();

  // Convert canvas image to base64 and resolve it
  var dataURL = canvas.toDataURL('image/jpeg', 0.8);
  google.colab.kernel.invokeFunction('notebook.handle_image', [dataURL], {});
}
</script>
"""

# Define a callback function to handle the captured image
def handle_image(data_url):
    # Decode the base64 image data
    binary = b64decode(data_url.split(',')[1])

    # Save the image to a file
    with open('captured_photo.jpg', 'wb') as f:
        f.write(binary)
    return len(binary)

# Register the callback function
from google.colab import output
output.register_callback('notebook.handle_image', handle_image)

def take_photo(filename='captured_photo.jpg'):
    # Display the HTML and JavaScript for photo capture
    display(HTML(PHOTO_CAPTURE_HTML))
    # Wait for the image data to be returned from JavaScript
    eval_js("console.log('Waiting for image capture...')")
    # Save the image data to the specified filename
    print(f"Photo saved as {filename}")


```

```python
# Call the function to start the photo capture process
take_photo()
```
